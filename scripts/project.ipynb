```python
# ręczne k fold
from sklearn.model_selection import StratifiedKFold
seed = 7
np.random.seed(seed)


kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
cvscores = []

i = 1
for train, test in kfold.split(X, y):
    
    # create model
    model = Sequential()
    model.add(Dense(300, input_dim=len(df.columns)-1, activation=LeakyReLU()))
    model.add(Dropout(0.25))
    model.add(Dense(150, activation=LeakyReLU()))
    model.add(Dropout(0.25))
    model.add(Dense(100, activation=LeakyReLU()))
    model.add(Dropout(0.25))
    model.add(Dense(50, activation=LeakyReLU()))
    model.add(Dropout(0.25))
    model.add(Dense(10, activation=LeakyReLU()))
    model.add(Dropout(0.25))
    # model.add(Flatten())
    model.add(Dense(1, activation='sigmoid'))

    # Compile model
    model.compile('sgd', loss='binary_crossentropy', metrics=[metrics.AUC()])
#     model.compile('sgd', loss='mse', metrics=[metrics.AUC()])

#     print(X[train].shape)
#     print(y[train].shape)
    # Fit the model
    model.fit(X[train], y[train], epochs=12, batch_size=20)
    
    # evaluate the model
    scores = model.evaluate(X[test], y[test], verbose=0)
    print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
    
    cvscores.append(scores[1] * 100)
    print("%.2f%% (+/- %.2f%%)" % (np.mean(cvscores), np.std(cvscores)))
    
    #save model
    model_name = 'weights\\' + str(i) + '_weights_acc_' + str(round(scores[1]*100,2)) + '.h5'
    model.save_weights(model_name)
    i += 1
```
