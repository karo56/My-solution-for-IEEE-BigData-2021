```python
test = pd.read_csv('..\\data\\01_tabelaryczny\\test_data_tabular_format.csv')
train = pd.read_csv('..\\data\\01_tabelaryczny\\training_data_tabular_format.csv')

# Domination
# test = test.loc[test['mode'] == 'Domination'].reset_index(drop=True)
# train = train.loc[train['mode'] == 'Domination'].reset_index(drop=True)

# Devastator
# test = test.loc[test['mode'] == 'Devastator'].reset_index(drop=True)
# train = train.loc[train['mode'] == 'Devastator'].reset_index(drop=True)


buff = train['winner']
train = train.drop(['winner'], axis=1)

test['jakie'] = 'test'
train['jakie'] = 'train'


train_and_test = pd.concat([train, test], axis=0).reset_index(drop=True)
train_and_test = train_and_test.drop(['t','id','players.0.player_id', 'players.1.player_id'], axis=1)


#kategorie
df_object = train_and_test.select_dtypes(include=['object'])
df_object = df_object.fillna('was_empty')
df_object = pd.get_dummies(df_object)

#numerki
df_without_object = train_and_test.select_dtypes(exclude=['object'])
df_without_object = df_without_object.fillna(-1)
# df_without_object = df_without_object.fillna(0)
# df_without_object = df_without_object.reset_index(drop=True)

# dfcolumns = df_without_object.columns
# for name in dfcolumns:
#     df_without_object[name].fillna(df_without_object[name].median(), inplace=True)

all_df = pd.concat([df_object, df_without_object], axis=1)

# train
all_df_train = all_df.loc[all_df['jakie_train'] == 1]
all_df_train = all_df_train.drop(['jakie_train','jakie_test'], axis=1)
all_df_train['winner'] = buff
# all_df_train.to_csv(r'nowe\\01_train.csv', index=False)

# test 
all_df_test = all_df.loc[all_df['jakie_test'] == 1]
all_df_test = all_df_test.drop(['jakie_train','jakie_test'], axis=1)
# all_df_test.to_csv(r'nowe\\01_test.csv', index=False)
all_df_test



# train_and_test
```

```python
from scipy.stats import ks_2samp

dfcolumns = list(all_df_test.columns)

p_values = []


for column in dfcolumns:
    x = all_df_train[column].values
    y = all_df_test[column].values
    w = ks_2samp(x, y)
    p_values.append(w[1])


# ks_2samp(x, y)
```

```python
all_p_values = pd.DataFrame(
    {'columns_name': dfcolumns,
     'p_value': p_values})

all_p_values = all_p_values.sort_values(by=['p_value'], ascending=True)

plt.figure(figsize=(18, 8))

plt.title("Wartość p value dla poszczególnych cech")
plt.plot(all_p_values['p_value'].values)
plt.ylabel('wartość')
plt.xlabel('numer zmiennej')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler


mms = MinMaxScaler()
mms.fit(all_df)
data_transformed = mms.transform(all_df)

```

```python
# k means
from sklearn.cluster import KMeans


Sum_of_squared_distances = []
K = range(1,40)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)
    
    
    
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
plt.savefig('kmeans.png')
```
